```
######## Maybe try to change the name of male, female into binary and location into number corresponding with 1, 2, 3, etc... ########
# Below we will split the dataset.csv into two sets. One set is training set, the other is the testing set.

# This will remove the entire column name "Surname, and CustomerId" with all of its values for every row.
# We removed this so our scalar.fit_transform() function below will only work with floats and not strings
X_Judge_Dataset = judge_dataset.drop(['Surname','CustomerId'], axis = 1) 
X_Judge_Dataset # Make sure all the data below is changed to numbers and that we got rid of Surname and CustomerId
```

```
# We will load the json file created our training-code.ipynb to get our training model into our test-code.ipynb
Nick_ANN_Model_File = open('/content/gdrive/My Drive/Colab Notebooks/CSCI 4931/Programming Assignment 1/Nick_ANN_Model.json', 'r')
Load_Nick_ANN_Model = Nick_ANN_Model_File.read()
Nick_ANN_Model_File.close()
Load_Nick_ANN_Model = model_from_json(Load_Nick_ANN_Model)
# We will load the h5 file created our training-code.ipynb to get our weights from our training model into our test-code.ipynb
Load_Nick_ANN_Model.load_weights("/content/gdrive/My Drive/Colab Notebooks/CSCI 4931/Programming Assignment 1/Nick_ANN_Model.h5")
print("Nick's ANN Model has been sucessfully loaded!") # Displays a sucessful message when the json and h5 file has been loaded.
```

```
# We need this to complete question 3 of programming assignment 1
Load_Nick_ANN_Model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy', 'Precision', 'Recall'])
```

```
# Making sure that my ANN model does not output all 0's or all 1's.

# Define min max scaler from the import file and save it to variable scaler
scaler = MinMaxScaler()
# Transform the data from X_Judge_Dataset
X_Judge_Test_Scaler = scaler.fit_transform(X_Judge_Dataset)

# We will use the ANN model to predict whether the customers exited or not on the judge.csv file.
Judge_Prediction = Load_Nick_ANN_Model.predict(X_Judge_Test_Scaler)
Get_Judge_Prediction = Judge_Prediction.round() # Rounds the predictions to 0 or 1. Before it would return only floats.
Get_Judge_Exit_Prediction = sum(Get_Judge_Prediction == 1) # Gets the total sum amount of 1's predicted.
Get_Judge_Not_Exit_Prediction = sum(Get_Judge_Prediction == 0) # Gets the total sum amount of 0's predicted.

# Prints the predicted total amount of people who exited and did not exited out of 1000.
# This is to ensure that we do not have predictions being all 0's or all 1's
print("The predicted amount of people exiting in judge.csv is:", Get_Judge_Exit_Prediction)
print("The predicted amount of people not exiting in judge.csv is:", Get_Judge_Not_Exit_Prediction)
```
